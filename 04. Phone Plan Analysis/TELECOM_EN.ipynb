- *** The best form to calculate the information is to define one function
- *** subtract the free package limit from the total number of calls, text messages, and data* #multiplique o resultado pelos valores do plano;
- ***  ADD THE PRICE/MONTH PER CALLS,MESSAGES,MB_USED)
- *** add subscription fee corresponding to rate plan To merge rates in a table.


````python
def count_for_pay(row):
#CALLS MINUTE
    if (row['minutes'] - row['minutes_included']) > 0:
        row['calls_for_pay'] = row['minutes'] - row['minutes_included']
    else:
        row['calls_for_pay'] = 0
    
## MESSAGES 

    if (row['messages'] - row['messages_included']) > 0:
        row['messages_for_pay'] = row['messages'] - row['messages_included']
    else:
        row['messages_for_pay'] = 0

## MB_USED1        
    if row['mb_used'] - row['mb_per_month_included'] > 0:
        row['mb_for_pay'] = row['mb_used'] - row['mb_per_month_included']
    else:
        row['mb_for_pay'] = 0
        
    return row

data_group = data_group.apply(count_for_pay, axis=1)
````

````python
data_group
````

````python
def bill(row):
        #CALLS_FOR_PAY
    if (row['calls_for_pay'] > 0 ):
        row['bill_calls'] = row['calls_for_pay'] * row['usd_per_minute']
    else:
        row['bill_calls'] = 0
    
        #MESSAGES_FOR_PAY
    if (row['messages_for_pay'] >0 ):
        row['bill_messages'] = row['messages_for_pay'] * row['usd_per_message']
    else:
        row['bill_messages'] = 0 
        
        #mb_for_pay: REMEMBER >> THE GB == 1024 MB
    if row['mb_for_pay'] >0:
        row['bill_mb'] = row['mb_for_pay']/1024 * row['usd_per_gb']
    else:
        row['bill_mb'] = 0
        
        
    return row

data_group = data_group.apply(bill, axis =1)
````

````python
data_group
````

````python
data_group['bill_total'] = data_group['usd_monthly_pay'] + data_group['bill_calls'] + data_group['bill_messages'] +data_group['bill_mb']
display(data_group)
````
